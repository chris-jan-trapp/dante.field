```python
with open('nodes_and_children.txt', 'r') as nas:
    lines = nas.readlines()
    
ids = list([l.strip().split()[-1] for l in lines])
ids
```

```python
children = []
nodes = []

for line in lines:
    if line.startswith('Node'):
        nodes.append(line.strip().split()[-1])
    if line.startswith('child'):
        children.append(line.strip().split()[-1])
children = list(set(children))
nodes = list(set(nodes))
len(children)
```
